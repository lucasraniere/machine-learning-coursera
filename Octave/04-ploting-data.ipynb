# __Ploting Data__

```octave
t = [0:0.01:0.98];
```

```octave
y1 = sin(2*pi*4*t);
```

```octave
graphics_toolkit ("gnuplot"); % ploting in jupyter notebook
plot(t, y1)
```

```octave
y2 = cos(2*pi*4*t);
```

```octave
plot(t, y2)
```

```octave
plot(t, y1);
hold on;
plot(t, y2, 'r');
xlabel('time')
ylabel('value')
legend('sin', 'cos')
title('My Plot')
print -dpng 'myPlot.png'
```

```octave
close
```

```octave
figure(1); plot(t, y1);
figure(2); plot(t, y2);
```

```octave
subplot(1,2,1) % divides plot a 1x2 grid, acces first element 
plot(t, y1)
subplot(1,2,2)
plot(t, y2)
axis([0.5 1 -1 1])
```
